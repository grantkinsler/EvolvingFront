```python
import numpy as np
import pandas as p
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
from adjustText import adjust_text
from Bio import SeqIO
from Bio.Seq import Seq
%matplotlib inline

sns.set_style('white')
sns.set_style('ticks')
sns.set_color_codes()
```

```python
pools = ['74D-1_ND',
'74D-1_041217',
'74D-2_112217',
'74D-6_112217',
'74D-7_112217',
'YP1_041217',
'YP1-B2-1_060217',
'YP10_031717',
'YP12_031717',
'YP4_031717']


yuping_estimate = {'74D-1_ND':18000,
'74D-1_041217':1300,
'74D-2_112217':32067,
'74D-6_112217':24222,
'74D-7_112217':16087,
'YP1_041217':12000,
'YP1-B2-1_060217':9000,
'YP10_031717':6300,
'YP12_031717':24000,
'YP4_031717':3200}
```

```python
def hamming_distance(s1, s2):
#     print(s1,s2)
    assert len(s1) == len(s2)
    return sum(ch1 != ch2 for ch1, ch2 in zip(s1, s2))
```

```python
pool_dict = {}

for pool in pools:
    pool_dict[pool] = p.read_table(f'cluster_final/{pool}_cluster_final.txt',names=['pair_name','high_diversity_BC','low_diversity_BC','count'])
```

```python
pool_dict
```

```python




for pool in pools:
    plt.figure()
    plt.title(pool)
    sns.distplot(pool_dict[pool]['count'].values,kde=False)
```

```python
cutoff = 80

x_list = []
y_list = []
mean_list = []
cutoff_mean_list = []
for pool in pools:
    print(pool,'\t',len(pool_dict[pool].index),'\t',len(pool_dict[pool][pool_dict[pool]['count']>30].values))
    m = np.mean(pool_dict[pool]['count'].values)
    print(m)
    cm = np.mean(pool_dict[pool][pool_dict[pool]['count']>cutoff]['count'].values)
    print(cm)
    mean_list.append(m)
    cutoff_mean_list.append(cm)
    x = yuping_estimate[pool]
    y = len(pool_dict[pool][pool_dict[pool]['count']>cutoff].values)
    x_list.append(x)
    y_list.append(y)
    
    if '74D' in pool:
        plt.scatter(x,y,color='b')
    else:
        plt.scatter(x,y,color='r')
        
texts = [plt.text(x_list[i]-0.1,y_list[i]+0.1,pools[i].split('_')[0]) for i in range(len(pools))]
        
plt.xlabel("Yuping's Plate Estimate")
plt.ylabel(f'Number of BC clusters with > {cutoff} reads')

adjust_text(texts)
plt.plot([0,1.1*max(max(x_list),max(y_list))],[0,1.1*max(max(x_list),max(y_list))],'k--')
plt.title('Diversity Estimate Comparison')
plt.savefig(f'diversity_est_comparison_{cutoff}readcutoff.png',dpi=200,bbox_inches='tight')
```

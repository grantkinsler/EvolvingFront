```python
input_filename = 'diversitycheck.inp'

with open(input_filename,'w') as input_file:

    for index in range(len(primer_assignments.index)):
        s_primer = primer_assignments['S_primer'].values[index]
        n_primer = primer_assignments['N_primer'].values[index]
        f_primer = primer_assignments['F_primer'].values[index]
        r_primer = primer_assignments['R_primer'].values[index]
        sample_name = primer_assignments['sample_name'].values[index]
        sample_counter = primer_assignments['sample_counter'].values[index]

        multiplex_file = f'diversitycheck_{sample_name}_multiplex.txt'
        f_fasta = f'18032FL-38-{s_primer}{n_primer}_{sample_counter}_L001_R1_001.fastq'
        r_fasta = f'18032FL-38-{s_primer}{n_primer}_{sample_counter}_L001_R2_001.fastq'

        f_seq = str(primer_set[f_primer].seq)[:6]
        print(f_seq)
        r_seq = str(primer_set[r_primer].seq)[8:]
        print(r_seq)

        with open(multiplex_file,'w') as multiplex:
            multiplex.write(f'{sample_name}\t0\t0\t{f_seq}\t{r_seq}')

        input_file.write(f'{f_fasta}\t{r_fasta}\t{multiplex_file}\n')

    
```

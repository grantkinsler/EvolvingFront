```python
import numpy as np
import pandas as p
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
# from adjustText import adjust_text
from Bio import SeqIO
from Bio.Seq import Seq
%matplotlib inline
import matplotlib.gridspec as gridspec
from matplotlib.backends.backend_pdf import PdfPages

import gzip
from cyvcf2 import VCF

from itertools import combinations

sns.set_style('white')
sns.set_style('ticks')
sns.set_color_codes()
```

```python
def draw_plate(ax,pick_me,destinations,plate):
    plt.title(plate)

    plt.pcolormesh(np.asarray(pick_me).reshape(8,12),cmap='Blues',vmin=0,vmax=3)
#     plt.colorbar()

#     called = np.asarray(called).reshape(8,12)

#     plate_ratio[plate] = np.sum(called)
    counter = 0
    for y in range(8):
        for x in range(12):
            if pick_me[y,x] == True:
#                 plt.scatter(x+0.5,y+0.5,marker='x',color='k',s=100)  
                plt.text(x+0.5,y+0.5,color='k',s=f'{destinations[counter]}',size=10,ha='center',va='center')
                counter +=1 


    plt.xticks([i+0.5 for i in range(12)],[i+1 for i in range(12)])
    plt.yticks([i+0.5 for i in range(8)],[chr(65+i) for i in range(8)])
    plt.gca().invert_yaxis()

    for i in range(12):
        plt.axvline(i,color='gray')
    for i in range(8):
        plt.axhline(i,color='gray')

    plt.xlim(0,12)
    plt.ylim(8,0)
```

```python
coverage = p.read_csv('coverage_ests.csv',header=None,names=['well','coverage'])

plates = ['EvolvingFront_WGS_Plate1','EvolvingFront_WGS_Plate2','EvolvingFront_WGS_Plate3','EvolvingFront_WGS_Plate4']

well_order = [f'{chr(65+r)}{c+1:02}' for r in range(8) for c in range(12)]

additions = {'well':[],'coverage':[]}

for plate in plates:
    for well in well_order:
        if f'{plate}_{well}' not in coverage['well'].values:
            print(f'{plate}_{well}')
            additions['well'].append(f'{plate}_{well}')
            additions['coverage'].append(0)

additions = p.DataFrame(additions)
coverage = p.concat([coverage,additions])

removal_wells = [f'EvolvingFront_WGS_Plate4_{well}'for well in well_order[60:]]

coverage = coverage[~coverage['well'].isin(removal_wells)]
```

```python
coverage = coverage.sort_values('well')

coverage[coverage['coverage']<30]['well'].values
```

```python
wells_to_pick = coverage[coverage['coverage']<30]['well'].values


source_wells = {}
destination_wells = {}

mapping_dict = {}

running_wells_chosen = 0

chosen_bcs = []

for plate in plates:
#     this_plate = this_set[this_set['plate']==plate]
    
    source_wells[f'{plate}'] = []
    destination_wells[f'{plate}'] = []
    
    for well in well_order:
        well_name = f'{plate}_{well}'
#         this_well = this_plate[this_plate['well']==f'{plate}-{well}']
        
        if well_name in wells_to_pick:
            source_wells[f'{plate}'].append(f'{well}')

            dest_plate_number = int(np.ceil((running_wells_chosen+1)/96))
            destination_wells[f'{plate}'].append(f'{well_order[(running_wells_chosen)%96]}')
            mapping_dict[f'{plate}-{well}'] = f'{well_order[(running_wells_chosen)%96]}'
            running_wells_chosen += 1

# p.DataFrame()
```

```python
with PdfPages('EvolvingFront_WGS_Plate5.pdf') as pdf:
    
    for p,plate in enumerate(plates):
        these_sources = source_wells[plate]
        pick_me = [True if entry in these_sources else False for entry in well_order]
        pick_me  = np.asarray(pick_me).reshape(8,12)

        destinations = destination_wells[f'{plate}']

        print(plate,p%2)
        if (p%2) == 0:
            fig  = plt.figure(figsize=(11*1,8.5*1))
        
            gs = gridspec.GridSpec(3, 2, figure=fig,height_ratios=[3.75,1,3.75],width_ratios=[12,12])
        
            this_ax = fig.add_subplot(gs[0])
        else:
            this_ax = fig.add_subplot(gs[1])
            
        draw_plate(this_ax,pick_me,destinations,plate)
        
        if (p%2) == 0:
#             fig  = plt.figure(figsize=(11,8.5))
        
#             gs = gridspec.GridSpec(2, 2, figure=fig)
        
            this_ax = fig.add_subplot(gs[4])
            
        else:
            this_ax = fig.add_subplot(gs[5])
            
        draw_plate(this_ax,pick_me,destinations,plate)
        
    #     [i+0.5 for i in range(12)]
#         plt.savefig(f'EvolvingFront_WGS_Plate5_{plate}.pdf')
        
        if (p%2) != 0:
            pdf.savefig()
    
### PRINT AT 110%!!!!
    
```

```python
def unit_vector(vector):
    """ Returns the unit vector of the vector.  """
    return vector / np.linalg.norm(vector)

def angle_between(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'::

            >>> angle_between((1, 0, 0), (0, 1, 0))
            1.5707963267948966
            >>> angle_between((1, 0, 0), (1, 0, 0))
            0.0
            >>> angle_between((1, 0, 0), (-1, 0, 0))
            3.141592653589793
    """
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)
    
    return np.arctan2(np.dot(np.cross(v1_u,v2_u),np.ones(len(v1_u))),np.dot(v1_u,v2_u))

def angle_between_array(v1,v2_array):
    
    angles = np.zeros(v2_array.shape[0])
    
    for entry in range(v2_array.shape[0]):
        angles[entry] = angle_between(v1,v2_array[entry,:])[0]
        
    return angles
#     return np.arccos(np.clip(np.dot(v1_u, v2_u), -1.0, 1.0))
```

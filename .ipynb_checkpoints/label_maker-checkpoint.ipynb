```python
import pandas as p
from datetime import date
import datetime
```

```python


evolutions = {'Evo1D':{'IRAs':['1','2'],'Tor1Cyr1Gpb2':['1','2']},
              'Evo2D':{'Ira1 MIS':['1','2'],'Ira1 NON':['1','2']},
              'Evo3D':{'Ira1 MIS':['1','2','3'],'Ira1 NON':['1','2','3'],'Tor1Cyr1Gpb2':['1','2']}  }


starting_date = date(2019,4,16)
n_transfers = 31

all_labels = []

for transfer in range(n_transfers):
    for evolution in evolutions.keys():
        for genotype in evolutions[evolution].keys():
            for replicate in evolutions[evolution][genotype]:
                

                if '1' in evolution:
                    this_d = starting_date + datetime.timedelta(days=1*transfer)
                elif '2' in evolution:
                    this_d = starting_date + datetime.timedelta(days=2*transfer)
                elif '3' in evolution:
                    this_d = starting_date + datetime.timedelta(days=3*transfer)

                all_labels.append([evolution,genotype,'Rep'+str(replicate),'T'+str(transfer),transfer,this_d.isoformat()])
#                 {'evolution':evolution,
#                                    'genotype':genotype,
#                                    'replicate':'Rep'+str(replicate),
#                                    'transfer':'T_'+str(transfer),
#                                    'date':this_d.isoformat()},ignore_index=True)
```

```python
label_df = p.DataFrame(all_labels,columns=['evolution','genotype','replicate','transfer','transfer_number','date'])
```

```python
label_df = label_df.sort_values(['date','evolution','transfer_number','genotype','replicate'])
```

```python
label_df = label_df.drop(['transfer_number'],axis=1)
```

```python
conditions = ['1D_Fitness_Li2019','2D_Fitness_Li2019','3D_Fitness_Li2019','5D_Fitness_Li2019']
color_list = []
for barcode in source_neutral_df['barcode'].values:
    
    if barcode in neutral_all_barcodes:
        color_list.append('r')
    else:
        color_list.append('b')
    

for x,y in combinations(conditions,2):

#     plt.figure(figsize=(4,4))
    plt.figure()
    plt.scatter(source_neutral_df[x].values,
            source_neutral_df[y].values,alpha=0.2,color=color_list)
    
    plt.xlabel(x)
    plt.ylabel(y)
    plt.gca().set_aspect('equal', adjustable='box')

# plt.scatter(li_neutral_df['3D_Fitness_Li2019'].values,
#             li_neutral_df['5D_Fitness_Li2019'].values,alpha=0.5)
```

```python
neutral_targets = neutral_all_barcodes 
these_neutrals = yuping_rearrays[yuping_rearrays['Barcode'].isin(neutral_targets)]


plates = []
n_unis = []
for plate in np.unique(these_neutrals['ReArranged_Plate'].values):
    print(plate)
    n_bcs = np.unique(these_neutrals[these_neutrals['ReArranged_Plate']==plate]['Barcode'].values)
    
    plates.append(plate)
    n_unis.append(len(n_bcs))

plates_sorted = np.asarray(plates)[np.argsort(n_unis)[::-1]]
```

```python
plates_sorted
```

```python
n_unis
```

```python
chosen_bc = []

plate_map = {}

for plate in plates_sorted[:4]:
    this_plate =  these_neutrals[these_neutrals['ReArranged_Plate']==plate]
    for bc in neutral_targets:
        
        if bc not in chosen_bc:
            if bc in this_plate['Barcode'].values:
                well = this_plate[this_plate['Barcode']==bc]['well'].values[0]
                plate_map[well] = bc
                chosen_bc.append(bc)
            
        


```

```python
well_order = [f'{chr(65+r)}{c}' for r in range(8) for c in range(1,13) ]
sorting_wells = {b: i for i, b in enumerate(well_order)}

plates = sorted(plates_sorted[:4])

running_wells_chosen = 0

bc_list = []
source_wells = {}
destination_wells = {}

source_wells_full = []
destination_wells_full = []

for plate in plates:

    source_wells[f'{plate}'] = []
    destination_wells[f'{plate}'] = []
#         destination_wells_full[f'{plate}'] = []

    for well in well_order:
        well_name = f'{plate}-{well}'
        if well_name in plate_map.keys():
            
#             source_wells[f'{plate}'].append()
            
            source_wells[f'{plate}'].append(f'{well}')
            destination_wells[f'{plate}'].append(well_order[(running_wells_chosen)%96])
            
            source_wells_full.append(f'{plate}-{well}')
            destination_wells_full.append(f'Li2019_neutrals_handpicked-{well_order[(running_wells_chosen)%96]}')

            bc =  plate_map[well_name]
            bc_list.append(bc)
            
            running_wells_chosen += 1
            
#             well_order[(running_wells_chosen)%96]  /

neutral_selection_df = p.DataFrame()

neutral_selection_df['source_well'] = source_wells_full
neutral_selection_df['destination_well'] = destination_wells_full
neutral_selection_df['barcode'] = bc_list

neutral_selection_df.to_csv('Handpicked_Li2019_neutrals_rearray.csv')
```

```python
def draw_plate(ax,pick_me,destinations,plate):
    plt.title(plate)

    plt.pcolormesh(np.asarray(pick_me).reshape(8,12),cmap='Blues',vmin=0,vmax=3)
#     plt.colorbar()

#     called = np.asarray(called).reshape(8,12)

#     plate_ratio[plate] = np.sum(called)
    counter = 0
    for y in range(8):
        for x in range(12):
            if pick_me[y,x] == True:
#                 plt.scatter(x+0.5,y+0.5,marker='x',color='k',s=100)  
                plt.text(x+0.5,y+0.5,color='k',s=f'{destinations[counter]}',size=10,ha='center',va='center')
                counter +=1 


    plt.xticks([i+0.5 for i in range(12)],[i+1 for i in range(12)])
    plt.yticks([i+0.5 for i in range(8)],[chr(65+i) for i in range(8)])
    plt.gca().invert_yaxis()

    for i in range(12):
        plt.axvline(i,color='gray')
    for i in range(8):
        plt.axhline(i,color='gray')

    plt.xlim(0,12)
    plt.ylim(8,0)
```

```python
with PdfPages(f'Handpicked_Li2019_neutrals_rearray.pdf') as pdf:
    
    fig = plt.figure(figsize=(11*1,8.5*1))
    for pl,plate in enumerate(plates):

        these_sources = source_wells[f'{plate}']

        pick_me = [True if entry in these_sources else False for entry in well_order]
        pick_me  = np.asarray(pick_me).reshape(8,12)

        destinations = destination_wells[f'{plate}']
        
        gs = gridspec.GridSpec(3, 2, figure=fig,height_ratios=[3.75,1,3.75],width_ratios=[12,12])
        
        if pl > 1:
            pl = pl + 2 
        this_ax = fig.add_subplot(gs[pl])

        draw_plate(this_ax,pick_me,destinations,plate)


    pdf.savefig()
```

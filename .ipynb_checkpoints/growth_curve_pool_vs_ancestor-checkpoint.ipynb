```python
import matplotlib.pyplot as plt
import pandas as p
import seaborn as sns
import numpy as np
%matplotlib inline
from scipy.stats import linregress
from scipy.spatial import distance
from matplotlib import colors
import scipy
import os
import copy
from matplotlib.backends.backend_pdf import PdfPages

from ast import literal_eval
from Bio import SeqIO
from itertools import combinations

import Levenshtein

sns.set_color_codes()
sns.set_style('white')
sns.set_style('ticks')


import FlowCytometryTools # package which allows you to analyze FACS data
```

```python
## ordered in the following way:
labels = ['Anc R1','Anc R2','Anc R3','50/50 R1','50/50 R2','50/50 R3','Pool R1','Pool R2','Pool R3']

data = [[0.04,0.04,0.04,0.06,0.06,0.07,0.08,0.08,0.09], #9am 7Apr
        [0.05,0.055,0.05,0.08,0.08,0.09,0.1,0.1,0.12], #10am 7Apr
        [0.08,0.07,0.07,0.11,0.1,0.1,0.095,0.095,0.12], #11am 7Apr
        [0.08,0.08,0.08,0.1,0.11,0.11,0.11,0.1,0.11], # 12pm 7Apr
        [0.10,0.11,0.11,0.10,0.12,0.12,0.10,0.11,0.11], #2pm 7Apr
        [0.13,0.12,0.13,0.12,0.11,0.12,0.12,0.12,0.12], #4:45 7Apr
        [0.18,0.17,0.16,0.20,0.19,0.18,0.18,0.18,0.20] # 12pm 8Apr
       ]

times = [16,17,18,19,21,24,43]
```

```python
data_swapped = np.asarray(data).swapaxes(0,1)

colors = [ ## all from ColorBrewer2 
'#bdd7e7',#light blue
'#6baed6',#medium blue
'#2171b5',#dark blue
'#cbc9e2',#light purple
'#9e9ac8',#medium purple
'#6a51a3',#dark purple
'#fcae91',#light red
'#fb6a4a',#medium red
'#cb181d']#dark red

labels = ['Anc R1','Anc R2','Anc R3','50/50 R1','50/50 R2','50/50 R3','Pool R1','Pool R2','Pool R3']
```

```python

# plotting each replicate separately (hard to see)
for flask in range(data_swapped.shape[0]):
     
    plt.plot(times,data_swapped[flask],color=colors[flask],alpha=0.7,marker='o',label=labels[flask])
plt.ylim(0.0,0.2)
plt.legend()
plt.axhline(y=0.12,color='k',linestyle='--',alpha=0.8)
# plt.yscale('log')
```

```python
condition_labels = ['Anc','50/50','Pool']

for condition in range(3):
        
    plt.errorbar(times,np.mean(data_swapped[condition*3:condition*3+3,:],axis=0),
                 yerr=2*np.std(data_swapped[condition*3:condition*3+3,:],axis=0),
                 color=colors[condition*3+2], # pick the darkest color we have for the condition
                 capsize=0.01,alpha=0.7,marker='o',label=condition_labels[condition])
plt.ylim(0.0,0.21)
plt.legend(loc='lower right')
# plt.axhline(y=0.12,color='k',linestyle='--',alpha=0.8)
# plt.yscale('log')
plt.ylabel('OD600')
plt.xlabel('Hours after Transfer')
plt.savefig('growth_curve_pool_vs_ancestor.pdf',bbox_inches='tight')
```

```python
file_prefix = 'growthcurvefacs/growth curve 2_Experiment_Group' ## 50ul (1:100 dilution), mixed 5x up and down
# file_prefix = 'growthcurvefacs/growth_curve_pool_vs_ancestor_Experiment_Group' ## 20ul (1:100 dilution), mixed 1x up and down

wells = [['B1','B2','B3','B4','B5','B6','B7','B8','B9'],
         ['C1','C2','C3','C4','C5','C6','C7','C8','C9'],
         ['D1','D2','D3','D4','D5','D6','D7','D9','D10'],
         ['E1','E2','E3','E4','E5','E6','E7','E8','E9'],
         ['F1','F2','F3','F4','F5','F6','F7','F8','F9'],
         ['G1','G2','G3','G4','G5','G6','G7','G8','G9'],
         ['H1','H2','H3','H4','H5','H6','H7','H8','H9']]


counts = []

samples = {}

for r,row in enumerate(wells):
    counts.append([])
    for entry in row:
        sample = FlowCytometryTools.FCMeasurement(ID=entry, datafile=f'{file_prefix}_{entry}.fcs')
        samples[entry] = sample
        counts[r].append(len(sample.data))
```

```python
condition_labels = ['Anc','50/50','Pool']

for condition in range(3):
        
    plt.errorbar(times,np.mean(counts_swapped[condition*3:condition*3+3,:],axis=0),yerr=2*np.std(counts_swapped[condition*3:condition*3+3,:],axis=0),color=colors[condition*3+2],
                 capsize=0.01,alpha=0.7,marker='o',label=condition_labels[condition])
# plt.ylim(0.0,0.21)
plt.legend(loc='lower right')
# plt.axhline(y=0.12,color='k',linestyle='--',alpha=0.8)
# plt.yscale('log')
plt.ylabel('FACS events')
plt.xlabel('Hours after Transfer')
# plt.savefig('growth_curve_pool_vs_ancestor.pdf',bbox_inches='tight')
```

```python
import numpy as np
import pandas as p
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors
# from adjustText import adjust_text
from Bio import SeqIO
from Bio.Seq import Seq
%matplotlib inline

import gzip
from cyvcf2 import VCF

from itertools import combinations

sns.set_style('white')
sns.set_style('ticks')
sns.set_color_codes()
```

```python
coverage = p.read_csv('coverage_ests.csv',header=None,names=['well','coverage'])

plates = ['EvolvingFront_WGS_Plate1','EvolvingFront_WGS_Plate2','EvolvingFront_WGS_Plate3','EvolvingFront_WGS_Plate4']

well_order = [f'{chr(65+r)}{c+1:02}' for r in range(8) for c in range(12)]

additions = {'well':[],'coverage':[]}

for plate in plates:
    for well in well_order:
        if f'{plate}_{well}' not in coverage['well'].values:
            print(f'{plate}_{well}')
            additions['well'].append(f'{plate}_{well}')
            additions['coverage'].append(0)

additions = p.DataFrame(additions)
coverage = p.concat([coverage,additions])

removal_wells = [f'EvolvingFront_WGS_Plate4_{well}'for well in well_order[60:]]

coverage = coverage[~coverage['well'].isin(removal_wells)]
```
